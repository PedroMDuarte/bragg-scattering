Suppose we have a true value of S(Q) and D, then the ratio I_MA / I_TOF is given by

```python
def It( S, D, s0, DW=0.81):
    det = 6.5
    c  = DW * 4. * det**2. / ( 4*det**2. + 2.*s0 ) 
    return ((S-1) + (c-1)*D/c) * c + 1 
```

```python
s0 = np.linspace( 0., 70., 100)
plot( s0, It(1.3, 0., s0 ) )
plot( s0, np.ones_like(s0)*1.3)
```
